```python
#Read the data of total population with 25+ age having no education for 2010 year

df_total_pop = pd.read_csv('https://raw.githubusercontent.com/jeevan0603/data690_jeevan/main/individual_project_final/total%20pop.csv')
df_total_pop.head()
```

```python
#Removing Unwanted columns

df_total_pop1 = df_total_pop.drop(columns=['Unnamed: 0','Country Code','Lending Type'])
df_total_pop1.head()
```

```python
 fig = px.bar(
    data_frame = df_total_pop1,
    title=" Bar Charts showing countries with no education of 25+ age",
    labels={"value":"% of total population with 25+ age and no education"},
    x="Country Name",
    y="value",
    color="Country Name"
)

fig.update_xaxes(categoryorder="total descending")

fig.show()
```

```python
#Read the data of total female population with 25+ age having no education for 2010 year

df_total_female_pop = pd.read_csv('https://raw.githubusercontent.com/jeevan0603/data690_jeevan/main/individual_project_final/female_pop.csv')
df_total_female_pop
```

```python
 fig = px.bar(
    data_frame = df_total_female_pop1,
    title="Bar Charts showing countries with no education of 25+ age",
    labels={"value":"% of female population with 25+ age and no education"},
    x="Country Name",
    y="value",
    color="Country Name"
)

fig.update_xaxes(categoryorder="total descending")

fig.show()
```

```python
fig = px.pie(
    df_total_pop1,
    values='value', 
    names='Country Name',
    title='Pie chart showing percentage of population with 25+ age and no education '
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
