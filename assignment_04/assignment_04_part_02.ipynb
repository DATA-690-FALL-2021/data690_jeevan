## Cleanse the data to make the year, population, cost, cost per person numerical type - use apply() function

```python
def clean(string):
  return string.replace("*","").replace(",","")
df["Year"] =df["Census Year"].apply(clean)
df["Total_pop"] = df["Total Population"].apply(clean)
df

def value(string):
  return string.replace(",","").replace("$","").replace(" Billion","000000000").replace("cents","")
df["Cost"] = df["Census Cost"].apply(value)
df["Avg_cost_per_person"] = df["Average Cost Per Person"].apply(value)
df
```
